```
def vectorizer(strx):
    global my_array
    global word_dictionary
    global ntwrd
    global count
    #lemmatizing and indexing
    token = []
    
    wnl = WordNetLemmatizer()
    for i in strx.split():
        if i not in stop_words and i != '':
            lemmat = wnl.lemmatize(i)
            x = word_dictionary.get(lemmat)
            if x:
                token.append(x)
            else:
                count +=1
                try:
                    word_dictionary[lemmat] = count
                    my_array[count] = model.wv[lemmat]
                    token.append(count)
                except:
                    ntwrd.append(lemmat)
                    count -= 1
    
    
    l = len(token)
    if l%10 == 0:
      pass
    else:
      l = l+10-(l%10)
    return pad_sequences([token],l)[0]
```

```
import random
def myGenerator(X,Y,length):
  while True:
    l = random.randint(0,length-1)
    yield X[l],Y[l]
```

```
from keras.initializers import Constant
def makeModel(genTrain,steps_per_epochs,genTest,validation_steps,epochs,embedding_dim,n_classes):
    global word_dictionary
    global my_array
    global count
    embd =  pd.DataFrame(my_array).values.T
    
    embM = Embedding(count,
                    embedding_dim,
                    weights = [embd],
                    trainable = False,
                    input_length = None
                    )    
    
    model = Sequential()
    model.add(embM)
    optm = Adam()
    model.add(Bidirectional(GRU(units = 128,dropout = .2,recurrent_dropout=.3,return_sequences = True)))
    
    model.add(GRU(units = 64,dropout = .2,recurrent_dropout=.3))
    model.add(Dense(n_classes,activation = 'softmax'))
    model.compile(loss = 'categorical_crossentropy',optimizer = optm,metrics = ['accuracy'])
    print(model.summary())
    model.fit_generator(genTrain, steps_per_epoch=steps_per_epochs, epochs=epochs, verbose=1, validation_data=genTest, validation_steps=validation_steps, max_queue_size=10,workers=2, use_multiprocessing=True)
  
    return model
    
```
